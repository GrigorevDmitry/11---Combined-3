### –ì–∏–ø–æ—Ç–µ–∑–∞ ‚Ññ1: –†–∞–∑–ª–∏—á–∏—è –¥–æ—Ö–æ–¥–∞ –º–µ–∂–¥—É –∫–ª–∏–µ–Ω—Ç–∞–º–∏

–î–ª—è –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É u-—Ç–µ—Å—Ç–æ–º –∏ t-—Ç–µ—Å—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –æ—Ü–µ–Ω–∏—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏–π –≤—ã–±–æ—Ä–æ–∫.

```python
alpha = .05
stat, p = st.levene(
    df.query('churn == 0')['estimated_salary'], 
    df.query('churn == 1')['estimated_salary']
)
print(f"p-–∑–Ω–∞—á–µ–Ω–∏–µ:{p}")
if p < alpha:
    print("–î–∏—Å–ø–µ—Ä—Å–∏–∏ –Ω–µ —Ä–∞–≤–Ω—ã")
else:
    print("–î–∏—Å–ø–µ—Ä—Å–∏–∏ —Ä–∞–≤–Ω—ã")
```

–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≥–∏–ø–æ—Ç–µ–∑—ã H0 –∏ H1:

    –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H0 - "–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π —Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç—ã —Ä–∞–∑–Ω—ã–º–∏"
    –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ù1 - "–ú–µ–∂–¥—É —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–∞—Ä–ø–ª–∞—Ç–∞–º–∏ –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞"

```python
alpha = .05

results = st.ttest_ind(
    df.query('churn == 0')['estimated_salary'], 
    df.query('churn == 1')['estimated_salary']
)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

    –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H0 - "–í–æ–∑—Ä–∞—Å—Ç –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"
    –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ù1 - "–í–æ–∑—Ä–∞—Å—Ç –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"

```python
alpha = .05

results = st.ttest_ind(
    df.query('churn == 0')['age'], 
    df.query('churn == 1')['age']
)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

### –ì–∏–ø–æ—Ç–µ–∑–∞ ‚Ññ3: –†–∞–∑–ª–∏—á–∏—è –ø–æ–ª–∞ –º–µ–∂–¥—É –∫–ª–∏–µ–Ω—Ç–∞–º–∏

–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≥–∏–ø–æ—Ç–µ–∑—ã H0 –∏ H1:

    –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H0 - "–ü–æ–ª –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"
    –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ù1 - "–ü–æ–ª –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"

```python
alpha = .05

results = st.ttest_ind(
    df.query('churn == 0')['female'], 
    df.query('churn == 1')['female']
)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≥–∏–ø–æ—Ç–µ–∑—ã H0 –∏ H1:

    –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H0 - "–ö—Ä–µ–¥–∏—Ç–Ω–∞—è –∫–∞—Ä—Ç–∞ –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"
    –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ù1 - "–ö—Ä–µ–¥–∏—Ç–Ω–∞—è –∫–∞—Ä—Ç–∞ –≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ç—Ç–æ–∫"

```python
alpha = .05

results = st.ttest_ind(
    df.query('churn == 0')['credit_card'], 
    df.query('churn == 1')['credit_card']
)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

### –í—ã–≤–æ–¥—ã.
